3. DataViz

```python
flights = sns.load_dataset("flights")

flights
```

```python
fig, axs = plt.subplots(ncols=2)
fig.set_size_inches(16, 5)


sns.set_style("whitegrid")
sns.lineplot(data=flights, x="year", y="passengers", ax=axs[0])
sns.swarmplot(data=flights, x="month", y="passengers", hue="year", ax=axs[1])
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=flights, x="month", y="passengers", hue="year", marker="o", markersize=8, lw=2)
```
